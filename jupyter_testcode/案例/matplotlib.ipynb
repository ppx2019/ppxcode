```python
#æ¨¡å—pyplotåŒ…å«å¾ˆå¤šç”Ÿæˆå›¾è¡¨çš„å‡½æ•°
import matplotlib.pyplot as plt
import numpy as np
input_values = [1,2,3,4,5,6]
squares = [1,4,9,16,25,36]
#plot()ç»˜åˆ¶æŠ˜çº¿å›¾
plt.plot(input_values,squares,linewidth=5)
#np.array()å°†åˆ—è¡¨è½¬æ¢ä¸ºå­˜å‚¨å•ä¸€æ•°æ®ç±»åž‹çš„å¤šç»´æ•°ç»„
x = np.array(input_values)
y = np.array(squares)
#annotate()ç»™æŠ˜çº¿ç‚¹è®¾ç½®åæ ‡å€¼
for a,b in zip(x,y):
    plt.annotate('(%s,%s)'%(a,b),xy=(a,b),xytext=(-20,10),
                 textcoords='offset points')
#è®¾ç½®æ ‡é¢˜
plt.title('Square Numbers',fontsize=24)
plt.xlabel('Value',fontsize=14)
plt.ylabel('Square of Value',fontsize=14)
#è®¾ç½®åˆ»åº¦çš„å¤§å°,bothä»£è¡¨xyåŒæ—¶è®¾ç½®
plt.tick_params(axis='both',labelsize=14)
#show()æ‰“å¼€matplotlibæŸ¥çœ‹å™¨ï¼Œå¹¶æ˜¾ç¤ºç»˜åˆ¶çš„å›¾å½¢
plt.show()
```
