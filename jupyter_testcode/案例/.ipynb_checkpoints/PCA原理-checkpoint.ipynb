```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
# PCA分析流程
X=np.array([[1,2],
            [2,4]])

# 1.样本中心化X，
X1=X-np.mean(X)
print("中心化后样本X1:",X1)

# 2.计算中心化后的样本的协方差矩阵，cov()计算协方差矩阵
X2=np.cov(X1)
print("样本协方差：",X2)

# 3.特征值分解：计算特征值和特征向量，eig()返回特征值和特征向量,特征向量的列为对应的特质值的向量
w,v=np.linalg.eig(X2)
print("特征值：{}\n特征向量：{}".format(w,v))

# 奇异值分解,u:特征向量，s:特征值(降序)，vh:
u,s,vh=np.linalg.svd(X2,full_matrices=True,hermitian=True)
u
s
vh
# print("svd：",u,s,vh)

# 4.将2维降为一维（中心化后数据与最大特征值矩阵做乘法）
# np.dot(X1,v[:,1])
```

```python
pca=PCA(n_components=1,copy=False)
# pca=PCA(n_components=0.98,copy=False)

# pca.fit(X1)
# 对样本进行降维，降维后结果使用正交向量表示原数据集
x22=pca.fit_transform(X)
pca.transform(X)
x22
# 查看保留多少信息
pca.explained_variance_ratio_.sum()
# 将降维后数据转为原数据
pca.inverse_transform(pca.transform(X))

# 保留的特征占所有特征的方差百分比，值越大代表保留信息量越多
print(pca.explained_variance_ratio_)

# 保留特征的方差
print(pca.explained_variance_)
print(pca.singular_values_)
```
