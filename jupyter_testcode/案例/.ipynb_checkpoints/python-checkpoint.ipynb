```python
# 离散出连续数据
a = np.array([0, 1, 1, 0, 1, 3, 6, 3, 0, 1])

# 筛选出大于0的数据的索引
pos = np.where(a > 0)[0]

# 求索引的差分，差分!=1后的数为不连续点; [0]取元祖第一个值，即不连续点的数组; +1,求差分以后的数组比原数组少1个，
split = np.where(np.diff(pos) != 1)[0] + 1
split

# 参数2：从该索引后切，a[pos]：根据筛选索引取出对应原值
arr = np.split(a[pos], split)
arr
```

```python
# 保留分为点
a = np.array([0, 0, 0, 1, 3, 4, 0, 1, 3, 4, 0])
b = np.where(np.diff(a != 0))[0] + 1
arr = np.split(a, b)
arr
```
